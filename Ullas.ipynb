```python
# Binary transform if cabin is present or not
titanic['cabin_null'] = np.where(titanic['Cabin'].isnull(),1,0)

# Find percent of missing cabin values for survivers and non survivers
titanic.groupby(['Survived'])['cabin_null'].mean()
```

```python
titanic = titanic[['Age','Fare','Survived']]
titanic.head()
```

```python
# Percentage of missing values
titanic.isnull().mean()
```

```python
# Function to impute with a fillna method
def impute_nan(titanic=titanic,variable='Age',strat=None, method=None):
    titanic[variable + "_" + method]=titanic[variable].fillna(strat)
```

```python
methods = [titanic.Age.mean(), titanic.Age.median(), titanic.Age.mode()]
names = ['mean', 'median', 'mode']

for m, name in zip(methods, names):
  type_ = int(m)
  impute_nan(titanic,'Age', type_, method=name)
```
